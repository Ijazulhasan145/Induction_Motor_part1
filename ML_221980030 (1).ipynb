```python
x_test_ar=x_test.to_numpy()
x_train_ar = x_train.to_numpy()
y_train_ar = y_train.to_numpy()
y_test_ar=y_test.to_numpy
```

# Euclidean distance:

```python
def euclidean(xtrain,xtest):
    d = np.sqrt(np.sum((xtrain-xtest)**2))
    return d
```

# k-NN Model Implementation:

```python
def KNN (x_test_k1,x_train_k1,y_train_k1,k):
    prediction = []
    for i in range(len(x_test_k1)):
        distances = []
        for j in range(len(x_train_k1)):
            dist = euclidean(x_test_k1[i], x_train_k1[j])
            distances.append((dist,x_train_k1[j],y_train_k1[j]))
        distances.sort(key=lambda x: x[0])
        neighbors = distances[:k]
        counter = Counter()

        for neighbor in neighbors:
            counter[neighbor[2][0]] += 1

        prediction.append(counter.most_common(1)[0][0])

    return prediction
```

```python
predictions = KNN(x_test_ar,x_train_ar,y_train_ar,2)
predictions
```
